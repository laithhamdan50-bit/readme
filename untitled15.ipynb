````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
from typing import Mapping
Mapping={'normal':0,'heavy':1,'high':2,'low':3}
data[ 'Traffic Situation ']=data['Traffic Situation'].replace(Mapping)
````

````python
from sklearn.preprocessing import LabelEncoder
LabelEncoder=LabelEncoder()
data['Traffic Situation']=LabelEncoder.fit_transform(data['Traffic Situation'])
````

````python
data
````

````python
data
````

````python
data['Day of the week'].value_counts()
````

````python
d={'Saturday':0,'Sunday':1,'Monday':2,'Tuesday':3,'Wednesday':4,'Thursday':5,'Friday':6}
data['Day of the week']=data['Day of the week'].replace(d)
````

````python
data['Day of the week']
````

````python
data['Time']=pd.to_datetime(data['Time'])
data['Hour']=data['Time'].dt.hour
data['Minute']=data['Time'].dt.minute
data.drop(axis=1,columns='Time',inplace=True)
````

````python
data
````

````python
for col in data.columns:
  plt.figure(figsize=(8,5))
  plt.hist(data[col].dropna(),bins=10,color='blue',edgecolor='red')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.show()
````

````python
plt.pie(data['Traffic Situation'].value_counts(),labels=data['Traffic Situation'].value_counts().index,autopct='%1.1f%%',colors=['red','blue','black','yellow'],explode=(0.1,0,0.1,0))
plt.title('traffic Situation distrib')
plt.show()
````

````python
data
````

````python
group=data[['CarCount','BikeCount','BusCount','TruckCount']].sum()
plt.figure(figsize=(10,6))
plt.pie(group,labels=group.index,autopct='%1.1f%%',colors=['red','blue','black','yellow'],explode=(0.1,0,0.1,0))
plt.title('vehicle type distribution')
plt.show()
````

````python
avg=data.groupby('Day of the week')['Total'].mean()
plt.figure(figsize=(8,5))
plt.bar(avg.index,avg.values,color='red')
plt.xlabel('Day of the week')
plt.ylabel('Average Total traffic of the day')
plt.show()
````
